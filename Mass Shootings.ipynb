```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as graph
import seaborn as sns
from IPython.display import display, HTML

%matplotlib inline
sns.set(font_scale=1.5)
sns.set_style('ticks')
```

```python
data = pd.read_csv('data/World Mass Shooting Stats.csv')
```

```python
sorted_data = data.sort_values('Total Mass Shooting Deaths', ascending=False)

graph.figure(figsize=(10,14))
graph.title('Number of Mass Shooting Deaths (2009-2013)')
sns.barplot(x=sorted_data['Total Mass Shooting Deaths'], y=sorted_data['Country'], orient='h', palette='deep')
graph.xlabel('Total Deaths')
graph.show()
```

Don't people just love graphs that look like this when it's about guns? How about we make 1 simple control... I also will leave the ordering in its previous place so you can see how much things have move around.

```python
graph.figure(figsize=(10,14))
graph.title('Deaths from Mass Shootings per 1,000,000 People (2009-2013)')
sns.barplot(x=sorted_data['Mass Shooting Deaths per 1e6'], y=sorted_data['Country'], orient='h', palette='deep')
graph.xlabel('Deaths per Million People')
graph.show()
```

Perhaps some people are just better shooters than other or some places are more crowded than others and there when a mass shooting does occur more people do. Well we have graphs for that also.

```python
graph.figure(figsize=(10,14))
graph.title('Mass Shootings Events per 1,000,000 People (2009-2013)')
sns.barplot(x=sorted_data['Mass Shooting Incidents per 1e6'], y=sorted_data['Country'], orient='h', palette='deep')
graph.xlabel('Incidents per 1 Million People')
graph.show()
```
