# Did Snape Die Too Young?

## Long story short... no.

“UK Male life-expectancy is __78.73y +/-15y__ (A USA number) means that __25.815%__ or __~1 in 4__ of people are expected to have died at or before this age. I crossed check my math with life insurance life tables (my math was assuming normal distribution). The difference between the life tables and my prediction was 0.185%. Snape may have died younger than average but he didn't die surprising young.”

```python
import numpy as np
import warnings
import seaborn as sns
from scipy.stats import norm
from matplotlib import pyplot as graph
from IPython.display import display, Markdown

%matplotlib inline

warnings.simplefilter('ignore')
sns.set(font_scale=1.5)
sns.set_style('whitegrid')
```

```python
life_expectancy = 78.73
expectancy_std = 15
age_list = np.linspace(40, 100, num=100)

life_distn = norm(life_expectancy, expectancy_std)
life_distn.mean = life_expectancy
life_distn.std = expectancy_std

death_function = lambda age: life_distn.cdf(age)

# Plot the probability of having already died given an age.
graph.plot(age_list, death_function(age_list), linewidth=3)
graph.title('CDF for Mortality')
graph.ylabel('Probability')
graph.xlabel('Age')
graph.show()

graph.plot(age_list, life_distn.pdf(age_list), linewidth=3)
graph.title('PDF for Mortality')
graph.ylabel('Probability Density')
graph.xlabel('Age')
graph.show()
```

```python
# Probability of Death at Snape's Age is...
display(Markdown('Probability of Death at Snape\'s Age was...\n**{}%**'.format(round(death_function(69), 5) * 100)))
```
