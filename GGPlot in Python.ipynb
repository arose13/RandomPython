# Aesthetics and Layers

I think this was the knocked down argument I personally had for wanting to accept R for atleast data exploration but since this Python's ggplot does _literally_ everything R's ggplot can do (and more since you can use any of Python's interaction and animation libraries) I lost that argument again. 

```python
# Raw Data
ggplot(aes(x='carat', y='price', color='clarity'), data=diamonds) + geom_point()
```

```python
# Controlling for scaling
ggplot(aes(x='carat', y='price', color='clarity'), data=diamonds) + geom_point() + scale_y_log() + scale_x_log()
```
