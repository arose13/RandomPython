# Web Scraping 

Let's play with beautiful soup for a sec!

Docs are here: https://www.crummy.com/software/BeautifulSoup/bs4/doc/

Dependencies:
- http://docs.python-guide.org/en/latest/scenarios/scrape/ (pip install lxml & requests)
- https://www.crummy.com/software/BeautifulSoup/ (pip install beautifulsoup4)

```python
import requests
from lxml import html
from bs4 import BeautifulSoup
from IPython.display import display, Markdown
```

```python
data = requests.get('https://www.crummy.com/software/BeautifulSoup/')
page = BeautifulSoup(data.text, 'html.parser')
```

## Try Selecting Things

```python
assert (page.head.title is page.title)

display(Markdown('## Title'))
print(page.head.title)
print(page.title)

display(Markdown('## Finding All'))
print(page.findAll('h1'))
print(page.findAll('li'))
```

# The Whole File Prettified

```python
print(page.prettify())
```
